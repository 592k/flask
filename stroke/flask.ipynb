```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/krc/Downloads/stroke/healthcare-dataset-stroke-data.csv', index_col=['id'])
```

```python
df2 = pd.read_csv('/Users/krc/Downloads/stroke/healthcare-dataset-stroke-data.csv', index_col=['id'])
```

```python
# 결측치 drop
df = df.dropna()
# gender : other 1
df = df.drop(df[df.gender=='Other'].index)
```

```python
df[df.dtypes[df.dtypes=='object'].index]
```

```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
```

```python
# gender : Male=1, Female=0, ever_married : Yes=1, No=0
df.loc[:,['gender','ever_married']] = df.loc[:,['gender','ever_married']].apply(le.fit_transform)
```

```python
df
```

```python
oe = preprocessing.OneHotEncoder(sparse=False)
```

```python
for col in ['work_type', 'Residence_type', 'smoking_status']:
    one = oe.fit_transform(df[[col]])
    df2 = pd.DataFrame(one, columns=oe.categories_[0], index=df.index)
    df = pd.concat([df,df2], axis=1)
    df.drop(columns=[col], inplace=True)
df
```

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD, Adam
from tensorflow.keras.callbacks import ModelCheckpoint, TensorBoard, ReduceLROnPlateau
```

```python
Label = df['stroke']
InputFeature = df.drop(columns=['stroke'])
```

```python
InputFeature.shape
```

```python
model = Sequential()
model.add(Dense(10, activation='linear', input_shape=(18,)))
model.add(Dense(10, activation='linear'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer=Adam(learning_rate=0.01), metrics=['accuracy'])
```

```python
model.summary()
```

```python
# Call-back 함수
# CheckPoint: Epoch 마다 validation 성능을 검증하여, best performance 일 경우 저장
CP = ModelCheckpoint(filepath='Models/{epoch:03d}-{loss:.4f}-{accuracy:.4f}.hdf5',
            monitor='loss', verbose=1, save_best_only=True, mode='min')

# Learning Rate 줄여나가기
LR = ReduceLROnPlateau(monitor='loss',factor=0.8,patience=3, verbose=1, min_lr=1e-8)

CALLBACK = [CP, LR]
```

```python
model.fit(x=InputFeature, y=Label, epochs=100, shuffle=True, batch_size=3, callbacks=CALLBACK)
```
